```python
#Step 6: MISSING VALUES 
```

```python
#Check missing values
df.isnull()
```

```python
#sum /count of missing value
df.isnull().sum()
```

```python
#Fill missing values
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce') # here age is converted to numeric because by default some str value was there 
                                                     #in that so we cant
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
df['Marks'] = df['Marks'].fillna(df['Marks'].median())
```

```python
df['City'] = df['City'].fillna('Unknown')
```

```python
# Drop missing values
df.dropna()
```

```python
#Step 7: SORTING & RANKING
#sorting
df.sort_values(by='Marks' , ascending=False)
```

```python
#ranking
#Rank
df["rank_m"]=df['Marks'].rank(ascending=False)
```

```python
#dense rank
df['rank']=df['Marks'].rank(method='dense',ascending=False)
print(df)
```
